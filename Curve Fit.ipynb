```python
from numpy import log, arange, array
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data
x = array([1, 2, 3, 4, 5, 10])
y = array([4530, 4661, 4736, 4781, 4809, 4872])

# Define the logarithmic function
def f(x,a,b,c):
    return a * log(x)**2 + b * log(x) + c

# Fit the logarithmic function to the data
(a,b,c), _ = curve_fit(f, x, y)

# Plot the data and the fitted function
plt.scatter(x, y, label='Data')
x = arange(1,25,0.1)
plt.plot(x, f(x,a,b,c),'--',label=f'{a:.1f}log(x) + {b:.1f}log(x) + {c:.1f}', color='red')
plt.xlabel('Discharge Duration at 40 V End Voltage (h)')
plt.ylabel('Energy capacity (Wh)')
plt.title('Polarium Datasheet vs Logarithmic Fit')
plt.legend()
plt.show()
```
